```python
# importing libraries
import numpy as np
import pandas as pd
```

```python
#Reading CASE-SHILLER Index into a dataframe
df_CS = pd.read_csv(r"C:\Users\S\Downloads\CASESHELLER.csv")

#Changing dtype of date column
df_CS["DATE"] = pd.to_datetime(df_CS["DATE"])

#Resetting Index
df_CS.reset_index(inplace = True)
df_CS.drop(columns = ["index"], inplace = True)

#Creating "Year" and "Month" columns
df_CS["Year"] = pd.DatetimeIndex(df_CS["DATE"]).year
df_CS["Month"] = pd.DatetimeIndex(df_CS["DATE"]).month
print(df_CS.shape)
df_CS.tail()
```

```python
# Reading Unemployment Rate Data into a dataframe
df_unemp = pd.read_csv(r"C:\Users\S\Downloads\UNEMP_RATE.csv", names = ["DATE", "UNRATE"],  skiprows = 1)
print(df_unemp.shape)
df_unemp.tail()
```

```python
# Reading Per Capita GDP Data into a dataframe
df_pcgdp = pd.read_csv(r"C:\Users\S\Downloads\PERCAPITA_GDP.csv", names = ["DATE", "Per_Capita_GDP"], skiprows = 1)
print(df_pcgdp.shape)
df_pcgdp.tail()
```

```python
# Reading Interest Rate Data into a dataframe
df_fed_rate = pd.read_csv(r"C:\Users\S\Downloads\INT_RATE.csv",names = ["DATE", "FEDFUNDS"],  skiprows = 1)
print(df_fed_rate.shape)
df_fed_rate.tail()
```

```python
#Reading Construction_price into a dataframe
df_cons_price_index = pd.read_csv(r"C:\Users\S\Downloads\CONSTRUCTION_PRICE.csv", names = ["DATE", "Cons_Materials"], skiprows = 1)
print(df_cons_price_index.shape)
df_cons_price_index.tail()
```

```python
#Reading  Consumer Price Index in a dataframe
df_CPI = pd.read_csv(r"C:\Users\S\Downloads\CPIAUCSL.csv", names = ["DATE", "CPI"], skiprows = 1)
print(df_CPI.shape)
df_CPI.tail()
```

```python
#Reading  Monthly new house supply in a dataframe
df_house = pd.read_csv(r"C:\Users\S\Downloads\MONTHLY_HOUSE_SUPPLY.csv", names = ["DATE", "Houses"], skiprows = 1)
print(df_house.shape)
df_house.tail()
```

```python
#Reading population above 65 in a dataframe
df_oldpop = pd.read_csv(r"C:\Users\S\Downloads\POPULATION_ABOVE_65.csv", names = ["DATE", "OLD_POPULATION"],skiprows=1)
print(df_oldpop.shape)
df_oldpop.tail()
```

```python
# Reading population percent in a dataframe
df_pop = pd.read_csv(r"C:\Users\S\Downloads\TOTAL_POPULATION.csv", names = ["DATE", "POPULATION"], skiprows=1)
print(df_pop.shape)
df_pop.tail()
```

```python
# Reading working age population in a dataframe
df_working_pop = pd.read_csv(r"C:\Users\S\Downloads\WORKING_AGE_POP.csv", names = ["DATE", "WORKING_AGE_POP"], skiprows = 1)
print(df_working_pop.shape)
df_working_pop.tail()
```

```python
# Reading property tax in a dataframe
df_property_tax = pd.read_csv(r"C:\Users\S\Downloads\PROPERTY_TAX.csv", names = ["DATE", "PROPERTY_TAX"], skiprows = 1)
print(df_property_tax.shape)
df_property_tax.tail()
```

```python
# Merging Per Capita GDP (Quarterly data)
df_pcgdp["DATE"] = pd.to_datetime(df_pcgdp["DATE"])
df_merged = pd.merge(df_CS,df_pcgdp, how = "left")
df_merged.head()
```

# The data is quarterly. We will impute for other months using linear interpolation after we create the final dataframe combining all the data.

```python
# Concating dataframes having monthly data to create one dataframe
df = pd.DataFrame()
df_bymonth = [df_merged, df_unemp,df_fed_rate,df_cons_price_index,df_CPI,df_working_pop,df_house]
for df1 in df_bymonth:
    df1["DATE"] = pd.to_datetime(df1["DATE"])
    df1 = df1.set_index("DATE")
    df = pd.concat([df,df1], axis = 1)
print(df.shape)
df.tail()
```

```python
print(df_oldpop.columns)
print(df_pop.columns)
print(df_property_tax.columns)
```

```python
# Merging other dataframes
others = [df_pop, df_oldpop, df_property_tax]
for df1 in others:
    if "Year" not in df1.columns:
        df1["Year"] = pd.DatetimeIndex(df1["DATE"]).year
        df1.set_index("DATE", inplace = True)
        df = pd.merge(df, df1, how = "left", on = "Year")
    else:
        df1.set_index("DATE", inplace = True)
        df = pd.merge(df, df1, how = "left", on = "Year")
df["DATE"] = df_CS["DATE"]
df.set_index("DATE", inplace = True)
df.head()

```
